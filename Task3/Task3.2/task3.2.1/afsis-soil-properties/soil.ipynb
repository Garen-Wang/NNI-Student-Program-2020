```python
train['Depth'] = train['Depth'].map(lambda x: 1 if x == 'Topsoil' else 0)
test['Depth'] = test['Depth'].map(lambda x: 1 if x == 'Topsoil' else 0)
train_x = train.drop(['PIDN', 'SOC', 'pH', 'Ca', 'P', 'Sand'], axis=1)
train_y = train[['Ca', 'P', 'pH', 'SOC', 'Sand']]
test_x = test.drop(['PIDN'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, BayesianRidge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, mean_squared_error, auc
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train_x, train_y.values, test_size=0.15, random_state=2077)
```

```python
# from sklearn.svm import SVR
# regressor = SVR(C=10000, verbose=2)

regressor = BayesianRidge()
pred = np.zeros((test_x.shape[0], 5))
for i in range(y_train.shape[1]):
    regressor.fit(X_train, y_train[:, i])
    pred[:, i] = regressor.predict(test_x)
pred
```

```python
df = pd.DataFrame(pred, columns=['Ca', 'P', 'pH', 'SOC', 'Sand'])
df.insert(0, 'PIDN', test['PIDN'])
```
